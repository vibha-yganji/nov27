```python
import pandas as pd
import numpy as np

mutations = pd.read_csv("mutations_file.csv")
mutations.head(5)
```

```python
import pandas as pd

filtered_columns = mutations.filter(like='ColoRect-AdenoCa', axis=1)

filtered_columns.head(5)
```

```python
filtered_columns['Row_Sum'] = filtered_columns.sum(axis=1)

summed = filtered_columns[['Row_Sum'] + [col for col in filtered_columns.columns if col != 'Row_Sum']]
summed.head(5)

```

```python

selected_columns = mutations[['Mutation type', 'Trinucleotide']]

sumy = pd.concat([selected_columns, summed], axis=1)

sumy.head(5)



```

```python
sumy.shape
```

```python
work = sumy.iloc[:, :3]
```

```python
import pandas as pd

group_column = 'Mutation type' 

sum_column = 'Row_Sum'  

# Group by the specified column and sum the values in the sum_column
grouped_df = sumy.groupby(group_column)[sum_column].sum().reset_index()

# Display the resulting DataFrame
print(grouped_df)
```

```python

gc = 'Trinucleotide' 

sc = 'Row_Sum'  

# Group by the specified column and sum the values in the sum_column
gdf = sumy.groupby(gc)[sc].sum().reset_index()



inserted_column = 'Mutation type'


print(gdf)
```
